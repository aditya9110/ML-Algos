## When to use PCA?
### - Data Visualization
### - Speeding ML Algorithm
### - Reducing Dimensionality or Features elements
### - Reduces the Model Complexity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
```

```python
cancer = datasets.load_breast_cancer()
```

```python
print(cancer.DESCR)
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
X_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2, random_state=42)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
```

```python
X_scaled.shape, X_pca.shape
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = cancer.target, cmap = 'viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
help(pca)
```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=8, random_state=42)
X_pca = pca.fit_transform(X_scaled)
variance = pca.explained_variance_ratio_
print(variance)

plt.bar(x = range(1, len(variance)+1), height=variance, width=0.7)
```

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
variance = pca.explained_variance_ratio_
print(variance)

plt.bar(x = range(1, len(variance)+1), height=variance, width=0.7)
```

```python
pca = PCA(n_components=20, random_state=42)
X_pca = pca.fit_transform(X_scaled)
variance = pca.explained_variance_ratio_
print(variance)

plt.bar(x = range(1, len(variance)+1), height=variance, width=0.7)
```
